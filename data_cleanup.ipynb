## Concatenate the dfs of the 5 months, remove rows with NaN values, then export df as csv

```python
df_all =  pd.concat([df_jan, df_feb, df_mar, df_apr, df_may], ignore_index=True)
df_all.tail()
```

```python
df_all['gender']
```

```python
df_all['Gender'] = df_all['gender'].replace([0,1,2],['unknown',"Male","Female"])
df_all = df_all.drop('gender',1)
df_all["distance"] = ''
df_all.tail(10)
```

```python
from vincenty import vincenty

def pandasVincenty(row):
    '''calculate distance (miles) between two lat&long points using the Vincenty formula '''

    return vincenty(['row.start station latitude, row.start station longitude'], ['row.end station latitude, row.end station longitude']).miles 
    
    df_all['distance'] =  df_all.apply(lambda r: pandasVincenty(r), axis=1)
    
    df_all.head(5)
```
